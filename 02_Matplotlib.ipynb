# Matplotlib

**Matplotlib** is a Python package for plotting graphs. It draws inspiration from the graphing functions in **MATLAB** and retains the names and arguments from similar functions in MATLAB. Matplotlib does not yet have the strong 3D graph plotting capabilities of MATLAB, but 2D plotting is quite capable. There are two predominant ways for plotting using Matplotlib:
* **pyplot** style
* **object-oriented** style

## Plotting using the pyplot style

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 501)
y = np.sin(x)
z = np.cos(x)
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, x, z)
plt.grid()
```

```python
plt.plot(x, y, x, z)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x versus y')
```

```python
plt.plot(x, y, x, z)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x versus y')
plt.legend(['sin(x)', 'cos(x)'])
```

```python
plt.plot(x, y, x, z)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x versus y')
plt.legend(['sin(x)', 'cos(x)'], loc=1)
```

```python
plt.plot(x, y, x, z)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x versus y')
plt.legend(['sin(x)', 'cos(x)'])
plt.savefig('fig01.png')
```

```python
plt.plot(x, y, x, z)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x versus y')
plt.legend(['sin(x)', 'cos(x)'])
plt.xlim(0, 4 * np.pi)
plt.ylim(-1.2, 1.2)
```

```python
def plot_graph(x, y,title='Graph'):
    plt.plot(x, y)
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.xlim(0, 4 * np.pi)
    plt.ylim(-1.2, 1.2)
    return

plot_graph(x, y, 'Graph of sin(x)')
```

```python
plot_graph(x, z, 'Graph of cos(x)')
```

```python
plt.subplot(1, 2, 1)
plot_graph(x, y, 'sin(x)')
plt.subplot(1, 2, 2)
plot_graph(x, z, 'cos(x)')
```

## Plotting using the object-oriented style

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$\sin (x)$')
ax.set_xlim([0, 4*np.pi])
ax.set_ylim([-1, 1])
xticks = [i*np.pi for i in range(5)]
ax.set_xticks(xticks)
yticks = np.arange(-1, 1.1, 0.5)
ax.set_yticks(yticks)
xtick_labels = [r'$0$', r'$\pi$'] + [r'${}\pi$'.format(k) for k in range(2, 5)]
ax.set_xticklabels(xtick_labels)
ytick_labels = np.arange(-1, 1.1, 0.5)
ax.set_yticklabels(ytick_labels)
ax.grid()
ax.plot(x, y)
```
